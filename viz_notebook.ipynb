```python
# load data
import pandas as pd
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 20)
##############################################################
pd_crime = pd.read_csv('Crime_Data_from_2020_to_Present_20240701.csv')
display(pd_crime.head())

##############################################################
url = "https://en.wikipedia.org/wiki/List_of_cities_in_Los_Angeles_County,_California"
tables = pd.read_html(url)
pd_pop = tables[0]
pd_pop = pd_pop[['City', 'Population as of (2020 Census)']].rename(columns={'Population as of (2020 Census)': 'Pop'})
```

```python
pd_crime['year'] = pd_crime['Date Rptd'].apply(lambda x: x[6:10])
year_filter = pd_crime['year'] == '2022'
pd_crime_2022 = pd_crime[year_filter]
```

# Temporal

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


pd_crime_2022['DATE OCC'] = pd.to_datetime(pd_crime_2022['DATE OCC'], format='%m/%d/%Y %I:%M:%S %p')
pd_crime_2022['Date Rptd'] = pd.to_datetime(pd_crime_2022['Date Rptd'], format='%m/%d/%Y %I:%M:%S %p')

pd_crime_2022 = pd_crime_2022[pd_crime_2022['DATE OCC'].dt.year == 2022]

pd_crime_2022['Month'] = pd_crime_2022['DATE OCC'].dt.month
monthly_crime = pd_crime_2022['Month'].value_counts().sort_index()


pd_crime_2022['DayOfWeek'] = pd_crime_2022['DATE OCC'].dt.dayofweek
daily_crime = pd_crime_2022['DayOfWeek'].value_counts().sort_index()



pd_crime_2022['Hour'] = pd_crime_2022['TIME OCC'].apply(lambda x: x // 100)
hourly_crime = pd_crime_2022['Hour'].value_counts().sort_index()

plt.figure(figsize=(10, 5))
sns.barplot(x=hourly_crime.index, y=hourly_crime.values, palette='viridis')
plt.title('Hourly Crime Trends in Los Angeles County (2022)')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Crimes')
plt.xticks(range(24))
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 5))
sns.lineplot(x=monthly_crime.index, y=monthly_crime.values,  color='skyblue')
plt.title('Monthly Crime Trends in Los Angeles County (2022)')
plt.xlabel('Month')
plt.ylabel('Number of Crimes')
plt.xticks(range(1, 13))
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 5))
sns.lineplot(x=daily_crime.index, y=daily_crime.values, color='coral')
plt.title('Daily Crime Trends in Los Angeles County (2022)')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Crimes')
plt.xticks(range(7), ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.grid(True)
plt.show()
```

```python
import folium
import pandas as pd
from folium import plugins

# Ensure you have valid latitude and longitude data for LA County
filtered_df = pd_crime_2022.dropna(subset=['LAT', 'LON'])
# Filter for coordinates within the LA County boundaries (approx.)
filtered_df = filtered_df[
    (filtered_df['LAT'] > 33.5) & (filtered_df['LAT'] < 34.8) &
    (filtered_df['LON'] > -119) & (filtered_df['LON'] < -117.5)
]


# Extract latitude and longitude into a list of tuples for the heatmap
crime_locations = list(zip(filtered_df['LAT'], filtered_df['LON']))

# Center the map on LA County 
la_county_center = [34.0522, -118.2437] 

# Create the base map
base_map = folium.Map(location=la_county_center, zoom_start=9) 

# Add the HeatMap layer
heatmap = plugins.HeatMap(crime_locations, radius=6, blur=2, gradient={'0':'Navy', '0.25':'Blue','0.5':'Green', '0.8':'Yellow','1': 'Red'})  
base_map.add_child(heatmap)

# Save the map as an HTML file
base_map.save("la_crime_heatmap.html")
```
